## РОЗНИЧНАЯ ТОРГОВЛЯ в ГЕРМАНИИ (мини-проект)

** Импорт набора данных из csv-файла в рабочем каталоге**

```python
import pandas as pd

# Запишите полученный датафрейм в retail
retail = pd.read_csv('germany.csv', encoding='ISO-8859-1')

retail
```

**Сохраните названия столбцов в переменной retail_columns**

```python
retail_columns = retail.columns

for i in retail_columns:
    print(i)
```

** Комментарии к столбцам набора данных:**
* InvoiceNo – идентификатор счета
* StockCode – идентификатор продукта
* Description – описание продукта
* Quantity – количество заказанных товаров
* InvoiceDate – дата транзакции
* UnitPrice – цена за шт.
* CustomerID – идентификатор клиента
* Country – страна проживания клиента.

**Проверьте, нет ли дубликатов. Выведите количество дубликатов (если есть)**

```python
retail.duplicated().sum()
```

**Удалить дубликаты (если таковые имеются)**

```python
retail = retail.drop_duplicates()

retail
```

**Подсчитать количество отмененных покупок**

```python
retail['InvoiceNo'].str.startswith('C').value_counts()
```

**Отфильтруйте набор данных. Должны остаться только запросы с количеством > 0**

```python
retail = retail.query('Quantity > 0')

retail.shape
```

**Подсчитайте количество покупок для каждого клиента из Германии. Распечатайте идентификаторы клиентов, которые попадают выше 80 процентилей.**

```python
germany_all = retail.query('Country == "Germany"') \
    .groupby(['CustomerID'], as_index=False) \
    .agg({'InvoiceNo': 'nunique'}) \
    .rename(columns = {'InvoiceNo': 'purchase_count'})

transactions_80 = germany_all.purchase_count.quantile(q=0.8)

germany_top = germany_all.query('purchase_count > @transactions_80').drop(columns='purchase_count') \
    .reset_index(drop=True)
```

```python
germany_all
```

```python
germany_top
```

**Сохраните идентификаторы таких клиентов в списке**

```python
lst = germany_top.CustomerID.to_list()

print(lst)
```

**Отфильтруйте исходный набор данных, оставив в нем только клиентов с идентификаторами из приведенного выше списка**

```python
top_retail_germany = retail.query('CustomerID in @germany_top.CustomerID')

top_retail_germany
```

**Найдите самый популярный продукт среди ведущих покупателей из Германии**

```python
top_retail_germany.groupby(['StockCode'], as_index=False) \
    .agg({'InvoiceNo': 'nunique'}) \
    .sort_values('InvoiceNo', ascending=False) \
    .rename(columns = {'InvoiceNo': 'count'})
```

**Подсчитайте выручку и добавьте эту информацию в соответствующий новый столбец.**

```python
retail = retail.assign(Revenue = retail.UnitPrice * retail.Quantity)

retail
```

**Подсчитайте общую стоимость каждой покупки. Перечислите идентификаторы пяти лучших покупок (с наибольшим общим доходом).**

```python
top5_inv = retail.groupby(['InvoiceNo'], as_index=False) \
    .agg({'Revenue': 'sum'}) \
    .sort_values('Revenue', ascending=False) \
    .head(5) \
    .InvoiceNo \
    .to_list()


', '.join(top5_inv)
```
