**Комментарии к столбцам набора данных Taxi Peru:**
* journey_id
* user_id
* driver_id
* taxi_id
* icon - тип поездки

* driver_start_lon – начальное местоположение водителя, долгота
* arrived_at – водитель вовремя прибыл в пункт выдачи

**Проверьте самую популярную платформу для оформления заказа (ответ в процентах)**

```python
taxi_source_percentage = round(100*(taxi.groupby('source').agg({'source': 'count'}) / taxi.journey_id.count()), 0)

taxi_source_percentage.loc[taxi_source_percentage.idxmax()]
```

**Рассчитайте распределение баллов водителей**

```python
driver_score_counts = taxi.groupby('driver_score', as_index=False) \
                    .agg({'journey_id': lambda x: round(x.count() * 100 / taxi['driver_score'].dropna().size, 2)}) \
                    .rename(columns={'journey_id': 'percentage'})\
                    .sort_values(by = 'percentage', ascending = False)

driver_score_counts
```

**Постройте график распределения баллов пилотов**

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


ax = sns.barplot(x='driver_score', y='percentage', data=driver_score_counts, color='blue', alpha=0.55 )
ax.set(xlabel='Driver score', ylabel='Percentage')
plt.show()
```

**Рассчитайте распределение баллов по пассажирам**

```python
rider_score_counts = taxi.rider_score \
    .value_counts(normalize=True) \
    .mul(100) \
    .round(2) \
    .reset_index()

rider_score_counts
```

**Постройте график распределения баллов пилотов**

```python
ax = sns.barplot(x='rider_score', y='proportion', data=rider_score_counts, color='red', alpha=0.55 )
ax.set(xlabel='Rider score', ylabel='Percentage')
plt.show()
```

**Отфильтруйте данные в наборе данных. Удалите все поездки, у которых start_type отличается от ASAP и которые ЗАРЕЗЕРВИРОВАНЫ**

```python
taxi = taxi.query('start_type == "asap" or start_type == "reserved"')

taxi.head()
```

**Показать количество заказов с указанием ASAP и RESERVED start_type**

```python
taxi.start_type.value_counts()
```

**Рассчитайте время задержки прибытия водителя для каждой поездки. Сохраните данные в новой колонке "ВРЕМЯ ОЖИДАНИЯ". Преобразуйте данные о ВРЕМЕНИ ОЖИДАНИЯ в минуты.**

```python
# расчет задержек
taxi['wait_time'] = taxi['arrived_at'] - taxi['start_at']

# преобразование чисел в минуты
taxi['wait_time'] = taxi['wait_time'].dt.total_seconds() / 60
```

```python
# проверка результата преобразования в минуты
taxi.wait_time
```

**Найдите водителей, опоздавших на встречу пассажира. Отсортируйте водителей от того, у которого наибольшее количество опозданий, до того, у которого наименьшее количество опозданий.**

```python
late_drivers = taxi.query('start_type == "reserved" and wait_time > 0') \
                   .groupby('driver_id') \
                   .agg({'journey_id': 'count'}) \
                   .sort_values('journey_id', ascending=False)
late_drivers
```

**Добавьте столбцы МЕСЯЦА и ДНЯ НЕДЕЛИ с соответствующими данными в наш фрейм данных.**

```python
taxi['month'] = taxi.start_at.dt.month
taxi['weekday'] = taxi.start_at.dt.strftime('%A')
```

```python
taxi[['month', 'weekday']]
```

**График распределения количества заказов за месяц**

```python
taxi_by_month = taxi.groupby(['month'], as_index=False) \
                    .agg({'journey_id': 'count'})

ax = sns.barplot(x='month', y='journey_id', data=taxi_by_month)
ax.set(ylabel='total rides');
```

**График количества заказов В ЗАВИСИМОСТИ от дня недели. Отсортируйте график по порядку дней недели (начиная с воскресенья). Постройте распределение start_type для каждого дня.**

```python
weekdays_order = ['Sunday',
                  'Monday',
                  'Tuesday',
                  'Wednesday',
                  'Thursday',
                  'Friday',
                  'Saturday']

plt.figure(figsize=(16, 12))
sns.countplot(x='weekday', data=taxi, order=weekdays_order, hue='start_type', palette=['#14213D', '#FCA311', 'red']);
plt.show()
```

**Рассчитать показатели DAU**

```python
taxi['day'] = taxi.start_at.dt.day
```

```python
# dau - daily active users
dau = taxi.groupby(['month', 'day'], as_index=False) \
      .agg({'user_id': 'nunique'}) \
      .rename(columns = {'user_id': 'DAU'})
```

```python
dau['day_of_year'] = range(1,366)

dau
```

**График DAU**

```python
plt.figure(figsize=(16, 12))
ax = sns.lineplot(x='day_of_year', y='DAU', data=dau, palette='inferno_r')
ax.set(ylabel='DAU')
plt.show()
```

**Рассчитать показатели MAU**

```python
# monthly active users
mau = taxi.groupby('month', as_index=False) \
          .agg({'user_id': 'nunique'}) \
          .rename(columns={'user_id': 'num_unique_users'})
```

**Линейный график MAU**

```python
xticks = range(1, 13)
plt.figure(figsize=(16, 12))
sns.lineplot(data=mau.num_unique_users, markers='o')
plt.xticks(xticks);
```

**Построение гистограммы MAU**

```python
plt.figure(figsize=(16, 12))
sns.barplot(x='month', y='num_unique_users', data=mau)
```
