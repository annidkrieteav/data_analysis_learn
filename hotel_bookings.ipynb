**Замена пробелов в названиях столбцов**

```python
def name_upd(name):
    return name.replace(' ', '_').lower()

bookings = bookings.rename(columns=name_upd)

bookings
```

**Топ-5 стран по бронированиям среди жителей**

```python
successful_bookings_top5 = bookings.query('is_canceled == 0') \
    .groupby('country') \
    .agg({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False).head() \
    .rename(columns={'is_canceled': 'successful bookings'})


successful_bookings_top5
```

**Среднее количество ночей проживания в зависимости от типа отеля.**

**Анализ планируемой даты заезда. Какой месяц был самым популярным в 2016 году? 2017?**

```python
for year in [2016, 2017]:
      max_is_canceled = bookings.query("arrival_date_year == @year and is_canceled == 0")\
            .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False)\
            .agg({'is_canceled': 'count'})
      message = "Самый популярный месяц в {}:"
      print(message.format(year), max_is_canceled['arrival_date_month']
            .loc[max_is_canceled.is_canceled.idxmax(axis=0)])
```

**Самый популярный месяц для отмены бронирования в 2015? 2016? 2017 годах?**

```python
import warnings
warnings.filterwarnings('ignore')

for year in [2015, 2016, 2017]:
    max_canceled_month = bookings.query('is_canceled == 1 and arrival_date_year == @year') \
    .groupby(['arrival_date_year', 'arrival_date_month']) \
    .agg({'is_canceled': 'count'}) \
    .idxmax()
    print(f'Большинство отмен бронирований в {year} произошло в {max_canceled_month[0][1]}.')
```

**Создайте новый столбец "total_kids", добавив в него количество детей и количество младенцев. У какого типа отелей наибольшее среднее значение total_kids?**

```python
# Создайте колонку total_kids, объединив столбцы children и babies.
# Для отелей какого типа среднее значение переменной оказалось наибольшим?
bookings['total_kids'] = bookings.children + bookings.babies

bookings.groupby('hotel', as_index=False) \
    .agg({'total_kids' : 'mean'}) \
    .round(2) \
    .max()
```

**Коэффициент оттока** - это скорость, с которой клиенты прекращают вести бизнес с организацией. Чаще всего он выражается как процент подписчиков услуг, которые прекращают свои подписки в течение заданного периода времени. Это также скорость, с которой сотрудники покидают свои рабочие места в течение определенного периода. Чтобы компания расширяла свою клиентуру, темпы ее роста (измеряемые количеством новых клиентов) должны превышать уровень оттока.   


**Рассчитайте коэффициент оттока для клиентов с детьми и клиентов, у которых детей нет.**

```python
# создание нового столбца
bookings['has_kids'] = bookings.total_kids > 0

# запрос количества случаев отмены бронирования для посетителей с детьми и без них
canceled_with_kids = bookings.query('has_kids == True and is_canceled == 1') \
    .is_canceled.value_counts()
canceled_no_kids = bookings.query('has_kids == False and is_canceled == 1') \
    .is_canceled.value_counts()

# итоговые значения для расчета скорости оттока
canceled_no_kids_int = canceled_no_kids[1]
canceled_with_kids_int = canceled_with_kids[1]
total_bookings_with_kids = len(bookings.query('has_kids == True'))
total_bookings_no_kids = len(bookings.query('has_kids == False'))

# функция для расчета скорости оттока
def churn_rate(total, canceled):
    return (canceled / total) * 100


print(f'Уровень оттока клиентов, у которых нет детей: {round(churn_rate(total_bookings_no_kids, canceled_no_kids_int), 2)}')
print(f'Уровень оттока клиентов с детьми: {round(churn_rate(total_bookings_with_kids, canceled_with_kids_int), 2)}')
```
